# CONTROLE DE DE CADASTROS DE CLIENTES E PRODUTOS

Tabela para cadastramento dos clientes - nome, endereço, telefone
Os clientes serão orgnizados por cidade, então criar Tabela cidade - cidade, estado
Criar Tabela produtos - descrição, quantidade em estoque, preco de custo e preço de venda
Tambem será cadastrado fabricante, para poder fazer o agrupamento do produto por fabricante - nome fabricante e site
Tabela vendas - poderá ser realizados varias vendas, para clientes diferentes e em datas diferentes. Portanto, será necessario armazenar para qual cliente a venda foi realizada em qual a data. Tambem será armazenado o valor total da venda e tambem o valor pago, para caso haja algum desconto. A venda será composta por produtos. Na venda poderá ser vendida quantidades de diferentes produtos. Então será criado uma Tabela auxiliar que será informado quais os produtos serão vendidos. Nessa tabela items terá os itens correspondente a venda, bem como a quantidade vendida de cada items.

O sistema irá controlar o cadastro de clientes e produdos, permitir categorizar os clientes por sua cidade, e os produtos pelo seu fabricante.
